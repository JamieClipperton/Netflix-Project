## Extract

```python
#Import dependencies
import os
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
#File Names
titles = "CSVfiles/titles.csv"
credits = "CSVfiles/credits.csv"
titles_df = pd.read_csv(titles)
credits_df = pd.read_csv(credits)
```

```python
titles_df.head()
```

titles_df.head()

```python
credits_df.head()
```

```python
#Merging csv files based off of id
netflix_data_df = pd.merge(titles_df, credits_df, on="id")
netflix_data_df
```

## Transform

```python
#Clean netflix_shows dataframe
#Remove movies and list only tv shows
netflix_shows_df = netflix_data_df.loc[netflix_data_df["type"] == "SHOW", :]
netflix_shows_df.head()
```

```python
# Merge dataframes and drop duplicates and remove columns
netflix_df = netflix_shows_df[['title','description','release_year','age_certification','runtime','genres','seasons','imdb_score','imdb_votes','name','character','role']].copy()
netflix_df = netflix_df.drop_duplicates(["title"])
netflix_df
```

## Load

```python
#Load dataframe into database
engine = create_engine('postgresql://postgres:Turtle25@localhost:5432/tvshows_db')
```

```python
netflix_df.to_sql(name='tvshows_db', con=engine, if_exists='append', index=False)
```

```python
#Confirming data
pd.read_sql_query('select * from tvshows_db', con=engine).head()
```
